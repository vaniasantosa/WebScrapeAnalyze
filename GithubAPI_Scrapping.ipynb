# 2. Extract the JSON corresponding to the first 100 contributors from this API

```python
#Following code in class

import warnings
import requests
import json
import codecs
from bs4 import BeautifulSoup

headers = {'User-agent': 'Mozilla/5.0'}

# fetch html and convert to json
url = 'https://api.github.com/repos/apache/hadoop/contributors?per_page=100'
response = requests.get(url, headers=headers)

contributors = response.json()
for contributor in contributors:
        login = contributor['login']
        contributions = contributor.get('contributions')
        print(f"Username: {login}, Contributions: {contributions}")
```

# 3.  For each of the 100 contributors extracted in (2), write code that accesses their user information and extracts "login", "id", "location", "email", "hireable", "bio", "twitter_username", "public_repos", "public_gists", "followers", "following", "created_at" (and print those to screen)

```python
#Following code in class

import warnings
import requests
import json
import codecs
from bs4 import BeautifulSoup


# Loop over each contributor and fetch user information
for contributor in contributors:
    user_url = contributor['url']  # The API url for the user's information is already part of the contributor's data
    user_response = requests.get(user_url, headers=headers)

    if user_response.status_code == 200:
        # Parse the user information from the response
        user_info = json.loads(user_response.text)

        # Extract the required fields
        login = user_info.get('login')
        user_id = user_info.get('id')
        location = user_info.get('location')
        email = user_info.get('email')
        hireable = user_info.get('hireable')
        bio = user_info.get('bio')
        twitter_username = user_info.get('twitter_username')
        public_repos = user_info.get('public_repos')
        public_gists = user_info.get('public_gists')
        followers = user_info.get('followers')
        following = user_info.get('following')
        created_at = user_info.get('created_at')

        # Print the extracted information
        print(f"Username: {login}")
        print(f"ID: {user_id}")
        print(f"Location: {location}")
        print(f"Email: {email}")
        print(f"Hireable: {hireable}")
        print(f"Bio: {bio}")
        print(f"Twitter Username: {twitter_username}")
        print(f"Public Repos: {public_repos}")
        print(f"Public Gists: {public_gists}")
        print(f"Followers: {followers}")
        print(f"Following: {following}")
        print(f"Account Created At: {created_at}")
        print("\n")  # Print a newline for better readability
```
